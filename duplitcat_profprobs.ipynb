```python
"""
Скрипт для получения списка учеников посетивших 2 и более профпроб
"""
```

```python
import pandas as pd
import openpyxl
from openpyxl.utils.dataframe import dataframe_to_rows
from openpyxl.utils import get_column_letter
import time
```

```python
path_to_end_folder = 'data'
```

```python
df = pd.read_excel('data/students.xlsx',skiprows=3)
```

```python
df.head()
```

```python
dupl_probs_df = df[df['кол-во посещенных профпроб в текущем году'] >=2]
```

```python
dupl_probs_df
```

```python
dupl_rows = [] # список для хранения строк
for index,row in dupl_probs_df.iterrows():
    name_probs = row['названия'].split(',') # названия проб
    date_probs = row['даты прохождения'].split(',') # даты прохождения
    for idx,name in enumerate(name_probs):
        new_row = row.copy() # копируем строку
        new_row['названия'] = name
        new_row['даты прохождения'] = date_probs[idx]
        dupl_rows.append(new_row)
# Создание нового датафрейма на основе новых строк
out_df = pd.DataFrame(dupl_rows)
```

```python
lst_out = ['Муниципалитет','ФИО','Дата рождения','образовательная организация','класс (без буквы)','буква класса',
           'кол-во посещенных профпроб в текущем году','названия','даты прохождения','ответственный педагог-навигатор']
```

```python
out_df = out_df[lst_out]
```

```python
out_df['даты прохождения'] = out_df['даты прохождения'].astype(str)
out_df['даты прохождения'] = out_df['даты прохождения'].apply(lambda x:x.strip())
```

```python
out_df['даты прохождения'] = pd.to_datetime(out_df['даты прохождения'],dayfirst=True,errors='ignore')
out_df = out_df.sort_values(by=['ФИО','даты прохождения'],ascending=True)
out_df['даты прохождения'] = out_df['даты прохождения'].dt.strftime('%d.%m.%Y')
```

```python
out_df['даты прохождения']
```

```python
t = time.localtime()
current_time = time.strftime('%H_%M_%S',t)
wb = openpyxl.Workbook()

for row in dataframe_to_rows(out_df,index=False,header=True):
    wb['Sheet'].append(row)
for column in wb['Sheet'].columns:
    max_length = 0
    column_name = get_column_letter(column[0].column)
    for cell in column:
        try:
            if len(str(cell.value)) > max_length:
                max_length = len(cell.value)
        except:
            pass
    adjusted_width = (max_length + 2)
    wb['Sheet'].column_dimensions[column_name].width = adjusted_width
wb.save(f'{path_to_end_folder}\Записанные на 2 и более профпроб {current_time}.xlsx')
```
