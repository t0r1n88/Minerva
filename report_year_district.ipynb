```python
import pandas as pd
import openpyxl
from openpyxl.utils.dataframe import dataframe_to_rows
from openpyxl.utils import get_column_letter
import time
```

```python
def count_not_zero(column):
    """
    Функция для подсчета количества значений отличающихся от нуля
    """
    lst = column.tolist() # превращаем в список
    print(lst)

    digt_lst = [value for value in lst if (isinstance(value,int) and value != 0)] # отбибраем числа не равные нулю
    return len(digt_lst)

    
    
```

```python
def count_verifi(column):
    """
    Функция для подсчета количества верифицированных школьников
    """
    lst = column.tolist() # превращаем в список
    verifi_lst = [value for value in lst if (isinstance(value,str) and value != 'Не верифицирован')] # отбибраем числа не равные нулю
    return len(verifi_lst)
    
```

```python
def count_rmg(column):
    """
    Функция для подсчета
    """
    lst = column.tolist() # превращаем в список

    str_lst = [value for value in lst if (isinstance(value,str) and 'Россия' in value)] # отбибраем числа не равные нулю
    return len(str_lst)
```

```python
df_temp_students = pd.read_excel('data/students.xlsx',skiprows=3)
df_temp_teachers = pd.read_excel('data/teachers.xlsx',skiprows=1)
```

```python
# Готовим колонки к обработке
df_temp_students['Дата последней активности'] = df_temp_students['Дата последней активности'].fillna('Не заполнено')
df__temp_teachers['Подтвердил участие в 2023 году'] = df__temp_teachers['Подтвердил участие в 2023 году'].fillna('Не заполнено')
df_temp_students[['количество пройденных диагностик∂','кол-во посещенных профпроб']] = df_temp_students[['количество пройденных диагностик∂','кол-во посещенных профпроб']].fillna(0)
```

```python



# Фильтруем учеников по заходившим в 2023
df_st = df_temp_students[df_temp_students['Дата последней активности'].str.contains('2023')]
df_teach = df__temp_teachers[df__temp_teachers['Подтвердил участие в 2023 году'] == 'Да']
```

```python
# Добавляем колонку для подсчета тех кто был в ёэтом году на профпробах
df_st['даты прохождения'] = df_st['даты прохождения'].fillna('Не заполнено')
df_st['даты прохождения'] = df_st['даты прохождения'].astype(str)
df_st['Профпроба 2023'] = df_st['даты прохождения'].apply(lambda x:1 if '2023' in x else 0 )
```

```python
out_df=pd.pivot_table(df_st,
              index=['Муниципалитет','образовательная организация','ответственный педагог-навигатор'],
              values=['ФИО','скан согласия родителей на обработку перс.данных','количество пройденных диагностик∂',
                     'Профпроба 2023','Статус верификации',
                      'названия уроков'],
              aggfunc={'ФИО':'count','скан согласия родителей на обработку перс.данных':'count',
                      'количество пройденных диагностик∂':count_not_zero,'Профпроба 2023':count_not_zero,
                      'Статус верификации':count_verifi,'названия уроков':count_rmg})
```

```python
out_df=out_df.merge(df_teach,left_on='ФИО педагога-навигатора',right_on='ФИО',how='outer')
```

```python
out_df.rename(columns={'ФИО':'Подтверждение участия в 2023'},inplace=True)
```

```python

out_df['Подтверждение участия в 2023'] = out_df['Подтверждение участия в 2023'].fillna('Не подтверждено')
out_df['Подтверждение участия в 2023'] = out_df['Подтверждение участия в 2023'].apply(lambda x:x if x=='Не подтверждено' else 'Подтверждено' )
```

```python
wb = openpyxl.Workbook()
```

```python
for row in dataframe_to_rows(out_df,index=False,header=True):
    wb['Sheet'].append(row)
# сохраняем по ширине колонок
for column in wb['Sheet'].columns:
    max_length = 0
    column_name = get_column_letter(column[0].column)
    for cell in column:
        try:
            if len(str(cell.value)) > max_length:
                max_length = len(cell.value)
        except:
            pass
    adjusted_width = (max_length + 2)
    wb['Sheet'].column_dimensions[column_name].width = adjusted_width
```

```python
t = time.localtime()
current_time =time.strftime('%H_%M_%S', t)
```

```python
wb.save(f'Отчет по педагогам-навигаторам {current_time}.xlsx')
```
